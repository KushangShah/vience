input = [[5, 6, 9, 2, 5],

````python
def get_common_elements(l1, l2):
    l2_copy = l2.copy()
    common_list = []
    for i in l1:
        if i in l2_copy:
            l2_copy.remove(i)
            common_list.append(i)

    return common_list
````

````python
input_list =[[5, 6, 9, 2, 5], 
      [5, 11, 2, 7, 8, 9, 6], 
      [5, 2, 8, 9, 6]]
````
