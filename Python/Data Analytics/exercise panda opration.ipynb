```python
# write the python program to verify if the givin input is armstrong number or not.

p = 0

num = input("enter numbere here: ")

a = [int(i) for i in num]

for j in range(len(a)):
  p = p + a[j]**len(a)

if p == int(num):
  print("This is the armstrong number.")
else:
  print("This is not a armstrong number.")
```
